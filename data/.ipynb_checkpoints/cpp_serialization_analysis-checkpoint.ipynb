```python
import pandas as pd
import matplotlib
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

# C++ Serialization Benchmark

```python
# Import data
dframe_cpp = pd.read_csv("cpp_data_nlohmann.csv")
dframe_cpp.head(15)
```

```python
# Time plot for serializers
x_ticks = range(10)
times = dframe_cpp["time"][:10].tolist()
labels = dframe_cpp["serializer"][:10].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: Serialization Times (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Time (ms)")
plt.bar(x_ticks, times, tick_label=labels)
```

```python
# Time plot for Nlohmann
x_ticks = range(5)
times = dframe_cpp["time"][10:].tolist()
labels = dframe_cpp["serializer"][10:].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: Nlohmann Serialization Times (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Time (ms)")
plt.bar(x_ticks, times, tick_label=labels)
```

```python
# Time plot for all
x_ticks = range(15)
times = dframe_cpp["time"].tolist()
labels = dframe_cpp["serializer"].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: All Serialization Times (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Time (ms)")
plt.bar(x_ticks, times, tick_label=labels)
```

```python
# Size plot for serializers
x_ticks = range(10)
size = dframe_cpp["size"][:10].tolist()
labels = dframe_cpp["serializer"][:10].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: Serialization Size (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Size (bytes)")
plt.bar(x_ticks, size, tick_label=labels)
```

```python
# Size plot for nlohmann
x_ticks = range(5)
size = dframe_cpp["size"][10:].tolist()
labels = dframe_cpp["serializer"][10:].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: Nlohmann Serialization Size (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Size (bytes)")
plt.bar(x_ticks, size, tick_label=labels)
```

```python
# Size plot for all
x_ticks = range(15)
size = dframe_cpp["size"].tolist()
labels = dframe_cpp["serializer"].tolist()
plt.xticks(rotation="vertical")
plt.title("C++: All Serialization Size (iter = {})".format(dframe_cpp["iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Size (bytes)")
plt.bar(x_ticks, size, tick_label=labels)
```

# Rust Serialization Benchmark

```python
# Import data
dframe_rs = pd.read_csv("rust_serialization_benchmark.csv")
dframe_rs
```

```python
# Time plot for serializers
x_ticks = range(7)
times = dframe_rs["Time"].tolist()
labels = dframe_rs["Name"].tolist()
plt.xticks(rotation="vertical")
plt.title("Rust: Serialization Times (iter = {})".format(dframe_rs["Iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Time (ms)")
plt.bar(x_ticks, times, tick_label=labels)
```

```python
# Size plot for all
x_ticks = range(7)
size = dframe_rs["Size"].tolist()
labels = dframe_rs["Name"].tolist()
plt.xticks(rotation="vertical")
plt.title("Rust: Serialization Size (iter = {})".format(dframe_rs["Iterations"][0]))
plt.xlabel("Serializer")
plt.ylabel("Size (bytes)")
plt.bar(x_ticks, size, tick_label=labels)
```
